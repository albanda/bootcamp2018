## Scipy

As you have seen during the lecture, Scipyis a package for high-level scientific computing. Its different submodules correspond to different applications, such as interpolation, integration, optimization, image processing, statistics, special functions, etc.

We are going to take a look at some of its features, but there is much more depending on what you might actually need.

### Optimisation

One of Scipy  submodules is `scipi.optimize` that allows an almost black-box set of optimisation algorithms. For a full guide check here https://www.scipy-lectures.org/advanced/mathematical_optimization/index.html.

We are going to see an example of how to fit a curve, obtaining the optimal parameters with a least squares.

```python
def f(t, omega, phi): # Function to be optimised
    return np.exp(omega * t + phi)

x = np.linspace(0, 20, 50)
y = f(x, -0.3, 3) + 1*np.random.normal(size=50)

popt, pcov = optimize.curve_fit(f, x, y)   
popt_b, pcov_b = optimize.curve_fit(f, x, y, bounds=(-2, [2, 2]))
# This second regression is bounded to [-2:2] intervals
```

```python
plt.plot(x, y, 'bo', label='data') # Blue dots

plt.plot(x, f(x, *popt), 'r-', label='fit: omega=%5.3f, phi=%5.3f' % tuple(popt))

plt.plot(x, f(x, *popt_b), 'g--',label='fit: omega=%5.3f, phi=%5.3f' % tuple(popt_b))

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

### Image processing 

Another submodule from scipy is the `scipi.ndimage` one. It contains various functions for multi-dimensional image processing. We are going to see how easy it is to apply histogram segmentation to a 2d image.

Histogram segmentation is a very simple way to extract features from a background. A 2d image can be seen as a matrix where each value corresponds to a color in the colormap. The idea is the following: 
the background is supposed to have a constant value (for example 0 +noise) that is distinguishable from the rest. Hence, plotting the histogram of the image values we should be able to see a peak corresponding to the background and using those values we could mask the image to obtain only the features we are interested into. 

```python
# Example of difference between the background and the foreground of an image
im = np.zeros((200,200))
im[10:100,10:50]=1
im=im+0.1*np.random.randn(200,200)

plt.figure(figsize=(11,4))
plt.subplot(121)
plt.imshow(im)

hist, bin_edges = np.histogram(im, bins=30)
bin_centers = 0.5*(bin_edges[:-1] + bin_edges[1:])

plt.subplot(122)
plt.plot(bin_centers, hist, lw=2)
```

```python
## Segmenting an image

from scipy import ndimage

# Code to create the image that needs later to be segmented
n = 10
l = 256
im = np.zeros((l, l))
np.random.seed(1)
points = l*np.random.random((2, n**2))
im[(points[0]).astype(np.int), (points[1]).astype(np.int)] = 1
im = ndimage.gaussian_filter(im, sigma=l/(4.*n)) #gaussian filter

mask = (im > im.mean()).astype(np.float)
mask += 0.1 * im
img = mask + 0.2*np.random.randn(*mask.shape)

hist, bin_edges = np.histogram(img, bins=60)
bin_centers = 0.5*(bin_edges[:-1] + bin_edges[1:])

binary_img = img > 0.5
```

```python
plt.figure(figsize=(11,4))

plt.subplot(131)
plt.imshow(img)
plt.axis('off')
plt.subplot(132)
plt.plot(bin_centers, hist, lw=2)
plt.axvline(0.5, color='r', ls='--', lw=2)
plt.text(0.57, 0.8, 'histogram', fontsize=20, transform = plt.gca().transAxes)
plt.yticks([])
plt.subplot(133)
plt.imshow(binary_img, cmap=plt.cm.gray, interpolation='nearest')
plt.axis('off')

plt.subplots_adjust(wspace=0.02, hspace=0.3, top=1, bottom=0.1, left=0, right=1)
plt.show()
```

```python
# Look at this
plt.figure()
g1=sns.distplot(h_array,kde=False,norm_hist=False)
g2=sns.distplot(r_array,kde=False,norm_hist=False)
g2=sns.distplot(a_array,kde=False,norm_hist=False)

fig,axes=plt.subplots(1,3, figsize=(20,5))
g1=sns.distplot(h_array, rug=True,ax=axes[0],color="blue")

g2=sns.distplot(r_array, rug=True,ax=axes[1],color="orange")
g3=sns.distplot(a_array, rug=True,ax=axes[2],color="green")

g1.set_xlabel("homicides counts")
g2.set_xlabel("rapes counts")
g3.set_xlabel("assaults counts")
```

```python
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
year=crime_table[crime_table["agency_code"]=="TX22701"]["report_year"].values
violent_crimes=crime_table[crime_table["agency_code"]=="TX22701"]["violent_crimes"].values


g1=sns.regplot(year,violent_crimes);
g1.set_xlabel("year")
g1.set_ylabel("violent crimes in Austin, TX")
```

```python
sns.jointplot(h_array[:1000,], a_array[:1000,], kind="hex", color="#4CB391")
```

But here is what seaborn is done for: look how powerful it is when combined with pandas tables.


```python
g1=sns.catplot(x="agency_code", y="violent_crimes", kind="box", data=crime_table,aspect=3);
for ax in g1.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(90)
```

```python
sub_data=crime_table.sort_values(by="agency_jurisdiction").iloc[0:200]
colors=len(sub_data.groupby("agency_jurisdiction"))
palette = sns.color_palette("rocket_r", colors)

sns.relplot(x="report_year", y="crimes_percapita",
            hue="agency_jurisdiction",
            palette=palette,
            height=5, aspect=.75, facet_kws=dict(sharex=False),
            kind="line", legend="full", data=sub_data)
```

```python
sub_data2=crime_table.sort_values(by="agency_jurisdiction").iloc[0:200]

pal = sns.cubehelix_palette(10, rot=-.25, light=.7)
g = sns.FacetGrid(sub_data2, row="agency_jurisdiction", hue="agency_jurisdiction", aspect=2, palette=pal)
g.map(sns.kdeplot, "violent_crimes", clip_on=False, shade=True, alpha=1)
```
